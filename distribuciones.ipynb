```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm # aka "te quiero demasiado" como nemotecnica
```

### Distribución Exponencial

```python
def dist_exponencial(ini, fin, v_lambda, delta): # se crea una función la cual sera llamada en el siguiente bloque solo con los valores
    x=np.arange(0, (1/v_lambda)*6, delta)
    ej_x=[]
    y_evento=[]
    y_normal=[]
    integral_normal=0
    for elemento in tqdm(x): #tqdm añade las barras de progreso y tiempos, ideal para deltas muy pequeñas
        y_elemento=delta*v_lambda*np.e**(-v_lambda*elemento) #calculo de la función.
        if elemento >= ini and elemento<=fin: # si el elemento perteneces al intervalo seleccionado 
            
            y_evento.append(y_elemento)
            ej_x.append(elemento)
            integral_normal+=y_elemento
            y_normal.append(0)
            
        else:  #guarda los elementos del complemento del evento.
            ej_x.append(elemento)
            y_evento.append(0) # p=0 para los intervalos que no son del complemento
            y_normal.append(y_elemento)
#Graficación-----------------------------------------------------------------------------------------------
    label1="Area de probabilidad del evento: " +str(round(integral_normal,4))
    label2="Probabilidad fuera del evento:   " +str(round(1-integral_normal,4))
    labels = [label1, label2]
    fig, ax = plt.subplots(figsize=(14, 5))
    ax.stackplot(ej_x, y_evento, y_normal, labels=labels)
    ax.legend(loc='upper right')
    

    return(integral_normal, ej_x, y_evento,y_normal) # guarda todos los datos obtenidos en esas variables, en caso de ocuparse para otros analisis
    
    
```

```python
#Ejemplo de los apuntes;          funcion_exponencial(inicial, final, lambda, precisión de intervalos)
valor_normal, eje_x, eje_y, y_normal=dist_exponencial(0,       1,        0.2, 0.0001)
```

## Distribución Normal

```python
def dist_normal(ini, fin, mu, sigma, delta):
    
    x=np.arange((-3.5*sigma+mu), (4*sigma+mu), sigma*delta)
    ej_x=[]
    y_evento=[]
    y_normal=[]
    integral_normal=0
    
    for elemento in tqdm(x): 
        y_elemento=(sigma*delta*np.e**(-(((elemento-mu)/sigma)**2)/2))/(sigma*((2*np.pi)**0.5))
        if elemento >= ini and elemento<=fin: 
            
            y_evento.append(y_elemento)
            ej_x.append(elemento)
            integral_normal+=y_elemento
            y_normal.append(0)
        else:
            ej_x.append(elemento)
            y_evento.append(0)
            y_normal.append(y_elemento)
    
#Graficación-----------------------------------------------------------------------------------------------    
    label1="Area de probabilidad del evento: " +str(round(integral_normal,4))
    label2="Probabilidad fuera del evento:   " +str(round(1-integral_normal,4))
    labels = [label1, label2]
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.stackplot(ej_x, y_evento, y_normal, labels=labels)
    ax.legend(loc='upper right') #localización de la leyenda 
    

    return(integral_normal, ej_x, y_evento,y_normal)
```

### Valido ejemplo de apuntes  


```python
##Valido ejemplo de apuntes       funcion_normal(inicio, fin, media, sigma, presicion)
valor_normal, eje_x, eje_y, y_normal=dist_normal(2000,  2400,  2000,   200, 0.0001)
```
